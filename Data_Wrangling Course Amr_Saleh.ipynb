```python
df_project = pd.DataFrame({
    'client':['kiwilytics','EY','KPMG','McKinsey'],
    'Budget_KUSD':[120,200,150,None]
})
print(df_project)
```

```python
# fill missing values whith mean
mean_budget = df_project['Budget_KUSD'].mean()
df_project=df_project.fillna(mean_budget)
print(mean_budget)
print(df_project)
```

```python
# Grouping & Aggregating DataFrames
df_revenue = pd.DataFrame({
    'company':['Kiwilytics','Pwc','EY','KPMG','Kiwilytics','Pwc'],
    'Quarter':['Q1','Q1','Q1','Q1','Q2','Q2'],
    'Revenue_kUSD':[120,200,180,150,130,210]
})
print(df_revenue)
groupped = df_revenue.groupby('company')['Revenue_kUSD'].sum()
print(groupped )
```

```python
# Merge
df_managers = pd.DataFrame({
    'ManagerID':[1,2,3],
    'Name':['Ola','Maher','Menna']
})
print(df_managers)
df_projects = pd.DataFrame({
    'projectID':[101,102,103],
    'ManagerID':[1,2,3],
    'Client':['Amazon','Sony','Apple']
})
print(df_projects)
merged_df=pd.merge(df_managers, df_projects, on='ManagerID',how='inner')
print(merged_df)
```

```python
# convert dates (Error Types: #coerce #ignore #raise)
df['orderdate'] = pd.to_datetime(df['orderdate'],errors = 'coerce')
#coerce--> NaNبيحول اى تاريخ مش عارف يحولو ل
```

```python
df['shippingcost'] = pd.to_numeric(df['shippingcost'],errors = 'coerce')
df.loc[df['shippingcost']<0,'shippingcost'] = np.nan
```

```python
# handle nulls
df['orderID'] = df['orderID'].fillna(method='ffill')#املأ اى قيمة null بالقيمة اللى قبلها
df['customerID']=df['customerID'].fillna("Unknown")
df['shipcity']=df['shipcity'].fillna("Unspecified")


```

```python
# standardize names
df['shipcountry']=df['shipcountry'].str.strip().str.title()#remove spaces and the frist char be capital
df['shipcity']=df['shipcity'].str.strip().str.title()
df['shippingCompany']=df['shippingCompany'].str.strip()
```

```python
# fix specific cases
df.loc[df['shippingCompany'].str.contains("Kiwilytics",na=False),'shoppingCompany']='Kiwilytics Goods Shipping LLC.'
```

```python
# Export cleaned file
df.to_csv("cleaned_orders_final.csv",index=False)
```
